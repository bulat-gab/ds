## Visualizing $(x_1,x_2,y)\quad y$-categorical

```python
n=100
X=vstack([randn(n,2),randn(n,2)+3])
Y=hstack([zeros(n),ones(n)])
```

```python
colors={0:'b', 1:'r'}
scatter(X[:,0], X[:,1], c=[colors[y] for y in Y], lw=0)
```

## Visualizing $(x_1,x_2,y,z),   z\in \mathbb{R}$

```python
n=100
X=vstack([randn(n,2),randn(n,2)+3])
Y=hstack([zeros(n),ones(n)])
Z=rand(200)**2
```

```python
colors={0:'b', 1:'r'}
scatter(X[:,0], X[:,1], c=[colors[y] for y in Y], s=5+30*Z, edgecolor='k')
```

## Visualizing $(x_1,x_2,y,z),   z\in \{c_1,c_2,...c_K\}$

```python
n=100
X=vstack([randn(n,2),randn(n,2)+3])
Y=hstack([zeros(n),ones(n)])
Z=rand(200)>0.2
```

```python
unique(Z)
```

```python
colors={0:'b', 1:'r'}
markers={0:'o', 1:'x'}
for z in unique(Z):
    XX = X[Z==z]
    YY = Y[Z==z]
    scatter(XX[:,0], XX[:,1], c=[colors[y] for y in YY], lw=1, s=30, marker=markers[z], edgecolor='k')
```

## Bar plots

```python
X=arange(10)
Y=X**2
bar(X,Y)
```

```python
barh(X,Y)
```

```python
barh(X,Y)
```

```python
colors={True:'r',False:'b'}
barh(X,Y,color=[colors[y>50] for y in Y])
```

```python
val = 3+10*rand(5)    # the bar lengths
pos = arange(5)+.5    # the bar centers on the y axis

figure(1)
barh(pos,val, align='center')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Accuracy')
title('Performance of methods')
grid(True)
```

```python
figure(2)
barh(pos,val, xerr=rand(5), ecolor='r', align='center')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Performance')
```

```python
val1 = 3+10*rand(5)    # the bar lengths
val2 = 2+10*rand(5)    # the bar lengths
pos = arange(5)    # the bar centers on the y axis
h=0.3

figure(figsize=[10,5])
before = barh(pos,val1, height=h, color='b')
after = barh(pos+h, val2, height=h, color='r')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Accuracy')
title('Performance of methods')

legend( (before, after), ('before transformation', 'after transformation') )

grid(True)
```

## Stacked bars

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = bar(ind, menMeans, width, color='r', yerr=menStd)
p2 = bar(ind, womenMeans, width, color='y',
             bottom=menMeans, yerr=womenStd)

ylabel('Scores')
title('Scores by group and gender')
xticks(ind + width/2., ('G1', 'G2', 'G3', 'G4', 'G5'))
yticks(np.arange(0, 81, 10))
legend((p1[0], p2[0]), ('Men', 'Women'))
```

## Displaying 2D functions

```python
def fun(x, y):
    return x**2+(y-3)**2
```

```python
n = 256
X = np.linspace(-10, 10, n)
Y = np.linspace(-6, 6, n)
#XX, YY = np.meshgrid(x, y)


F = zeros([n,n])
for (ix,x) in enumerate(X):
    for (iy,y) in enumerate(Y):
        F[iy,ix] = fun(x,y)  # look at order: iy, ix
        
plt.contourf(X, Y, F, 18, c='black', alpha=.75, cmap='hot')
colorbar()
C = plt.contour(X, Y, F, 18, colors='black', linewidth=.5)
xlabel('x')
ylabel('y')
```

# 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2+Y**2

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = randn(50)
Y = randn(50)
Z = randn(50)

ax.scatter(X, Y, Z)
```
