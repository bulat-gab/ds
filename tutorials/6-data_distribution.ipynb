# 2D data in 3D

```python
X=randn(300,2)
Y=ones(300, dtype=int)
X[:150,0]+=3
Y[:150]=2

A=array([[1,1,1],[1,1,-2]]).T
X = np.dot(A, X.T).T  # convert 2D to 3D

X+=0.2*vec(randn(300))*row(array([1,-1,0]))  # add small orhogonal random noise
```

```python
from common.visualize.data import pca_explained, pca_2D
```

```python
pca_explained(X)  # most of the data is explained by first 2 components, since it is almost 2D
```

```python
pca_2D(X, Y, task='classification')
```
